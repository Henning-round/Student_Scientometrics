```python
homeworks=[]
category=[]
with open('henning_germanistik_metadata.csv', encoding="utf-8") as f:
    csv_reader = csv.reader(f)
    count=0
    category=[]
    for line in csv_reader:
        if count==0:
            category=[x for x in line]
            category[1] = "category"
            category[2] = "Year of publishing"
            count+=1
            continue
        
        work={}
        for index,val in enumerate(line):
            work[category[index]]=val
        homeworks.append(work)
```

```python
works = load_obj("cleaned_corpus")
```

```python
urls= [x.link for x in works]

for index,w in enumerate(list(homeworks)):
    
    if w["link"] not in urls:
        homeworks.remove(w)
    if w["link"] == "https://www.hausarbeiten.de/document/16801":
        w["institution"] = "Technische Universität Berlin"
    if w["link"] == "https://www.hausarbeiten.de/document/379315":
        w["institution"] = "Universität Kassel"
    if w["link"] == "https://www.hausarbeiten.de/document/501982":
        w["institution"] = "Universität Leipzig"
    if w["link"] == "https://www.hausarbeiten.de/document/494152":
        w["institution"] = "Universität Potsdam"
```
